### **Variables** üì¶

  **¬øSer√°n como las variables en matem√°ticas?**


Piensa en las variables como cajas donde puedes guardar diferentes tipos de datos. Puedes etiquetar estas cajas con un nombre para recordarte qu√© tipo de datos has guardado en cada una. Por ejemplo, si tienes una caja etiquetada como "edad", puedes guardar un n√∫mero en ella que represente la edad de alguien.

En Python, una variable se define simplemente dando un nombre a la variable y asign√°ndole un valor usando el signo =. Por ejemplo:
```
    edad = 20  # Aqu√≠, la variable "edad" guarda el n√∫mero 20
    nombre = "Ana"  # La variable "nombre" guarda el texto "Ana"
```

**Tipos de datos en Python:**


Python tiene varios tipos de datos que puedes guardar en las variables. Aqu√≠ te explico los m√°s comunes:

*   Strings (Cadenas de texto):\
Los strings son textos o cadenas de caracteres. Se representan con comillas simples ' o comillas dobles ".

    Ejemplo:
```
    saludo = "Hola"  # Un string que contiene la palabra "Hola"
```

*   N√∫meros (int, float, ...):\
    Python maneja diferentes tipos de n√∫meros:

    Enteros (int): Son n√∫meros enteros sin decimales.
    Ejemplo:
```
    edad = 25  # Un n√∫mero entero
```
    Flotantes (float): Son n√∫meros con decimales.
    Ejemplo:
```
    precio = 19.99  # Un n√∫mero con decimales
```






````python
# con el s√≠mbolo # podemos hacer comentarios que Python no leer√°
x = 5
print(type(x))  # Muestra <class 'int'>, indicando que x es un entero

y = "Hola"
print(type(y))  # Muestra <class 'str'>, indicando que y es un string
````

### **Funciones B√°sicas** üëæ

**'print' en Python:**

La funci√≥n print es una funci√≥n muy com√∫n en Python que se usa para **mostrar informaci√≥n en la pantalla**. Puedes usarla para mostrar mensajes, valores de variables o cualquier otra informaci√≥n que quieras que el usuario vea.

````python
#Mostrar mensajes
print("¬°Hola, mundo!")
````

````python
#Mostrar el valor de una Variable
edad = 20
print("Edad:", edad)
````

````python
#Mostrar varios valores a la vez
nombre = "Ana"
edad = 25
print("Nombre:", nombre, "Edad:", edad)
````

**'input' en Python:**

La funci√≥n input se usa para **pedir datos al usuario**. Cuando se ejecuta input, se detiene el programa hasta que el usuario ingrese alg√∫n dato y presione Enter. **El valor que el usuario ingresa se guarda como un string.**

````python
#Pedir un string al usuario
nombre = input("Por favor, ingresa tu nombre: ")
print("Hola, " + nombre + "!")
````

````python
#Pedir un n√∫mero al usuario
edad = input("Por favor, ingresa tu edad: ")
edad = int(edad)  # Convierte el string a entero
print("Tienes", edad, "a√±os.")
````

### **Condicionales** ‚úîÔ∏è‚ùå



*    **if** y **else**\
Los condicionales en Python te permiten tomar decisiones basadas en condiciones espec√≠ficas. Puedes usar la palabra clave **if** para evaluar una condici√≥n y ejecutar un bloque de c√≥digo si la condici√≥n es verdadera ‚úîÔ∏è.\
Tambi√©n puedes usar **else** para ejecutar un bloque de c√≥digo si la condici√≥n de if es falsa ‚ùå.

```
# Sintaxis b√°sica de un condicional
if condici√≥n:
    # Bloque de c√≥digo si la condici√≥n es verdadera
else:
    # Bloque de c√≥digo si la condici√≥n es falsa

```

**Ejemplo Condicionales (if - else)**

````python
edad = int(input("¬øQu√© edad tienes?"))

if edad >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
````
