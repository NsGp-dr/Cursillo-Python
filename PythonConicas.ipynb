# **Python y las secciones Cónicas** 👀🤯


¿Alguna vez te has preguntado cómo los **matemáticos** y **científicos** **describen las formas** que vemos todos los días, como los **círculos**, **óvalos** y **parábolas**? Esas formas especiales se llaman **secciones cónicas** y *se obtienen cortando un cono con un plano de diferentes formas*. Imagina que tienes un cono (como el que se usa para hacer helados) y lo cortas con un cuchillo de diferentes maneras:

conicas.png


*    **Círculo**: Si cortas el cono de forma horizontal, obtendrás un círculo perfecto, como el sol o una pelota.

*    **Elipse**: Si cortas el cono en un ángulo, obtendrás una elipse, que es como un óvalo. Los planetas siguen órbitas elípticas alrededor del sol.

*    **Parábola**: Si cortas el cono en un ángulo paralelo a uno de sus lados, obtendrás una parábola, que es como la trayectoria que sigue una pelota lanzada o el reflejo de una antena parabólica.

*    **Hipérbola**: Si cortas el cono verticalmente, obtendrás dos curvas abiertas llamadas hipérbolas. Estas se ven en cosas como las órbitas de algunas naves espaciales

Las secciones cónicas tienen aplicaciones increíbles en la vida real, desde diseñar lentes de cámaras hasta calcular trayectorias de cohetes. ¡Explorar las secciones cónicas te ayudará a entender cómo funciona el mundo que te rodea de una forma divertida y emocionante!

## **Las Cónicas en Python**

Ya sabemos que python es un lenguaje de programación bastante útil, ahora vamos a ver cómo podemos usarlo para visualizar dichas secciones. 👾

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la figura y los ejes
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# 1. Círculo: x^2 + y^2 = r^2
r = 3
theta = np.linspace(0, 2 * np.pi, 100)
x_c = r * np.cos(theta)
y_c = r * np.sin(theta)
axs[0, 0].plot(x_c, y_c)
axs[0, 0].set_title("Círculo: $x^2 + y^2 = r^2$")
axs[0, 0].set_aspect('equal')

# 2. Elipse: x^2 / a^2 + y^2 / b^2 = 1
a, b = 4, 2
theta = np.linspace(0, 2 * np.pi, 100)
x_e = a * np.cos(theta)
y_e = b * np.sin(theta)
axs[0, 1].plot(x_e, y_e)
axs[0, 1].set_title(f"Elipse: $\\frac{{x^2}}{{{a}^2}} + \\frac{{y^2}}{{{b}^2}} = 1$")
axs[0, 1].set_aspect('equal')

# 3. Parábola: y = a * x^2 + b * x + c
a_p, b_p, c_p = 1, 0, 0
x_p = np.linspace(-5, 5, 100)
y_p = a_p * x_p**2 + b_p * x_p + c_p
axs[1, 0].plot(x_p, y_p)
axs[1, 0].set_title(f"Parábola: $y = {a_p}x^2 + {b_p}x + {c_p}$")

# 4. Hipérbola: x^2 / a^2 - y^2 / b^2 = 1
a_h, b_h = 2, 1
x_h = np.linspace(-5, 5, 100)
y_h_positive = np.sqrt((x_h**2 / a_h**2 - 1) * b_h**2)
y_h_negative = -np.sqrt((x_h**2 / a_h**2 - 1) * b_h**2)
axs[1, 1].plot(x_h, y_h_positive)
axs[1, 1].plot(x_h, y_h_negative)
axs[1, 1].set_title(f"Hipérbola: $\\frac{{x^2}}{{{a_h}^2}} - \\frac{{y^2}}{{{b_h}^2}} = 1$")
axs[1, 1].set_xlim(-5, 5)
axs[1, 1].set_ylim(-5, 5)

# Configuramos los ejes y mostramos el gráfico
for ax in axs.flat:
    ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
    ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
    ax.grid(True, which='both')

plt.tight_layout()
plt.show()
```

Como sabemos, en python podemos programar casi de todo y de miles de formas, por lo cuál, esta es la forma más básica de visualizar datos de cualquier tipo en python, usando la biblioteca **matplotlib**.

Ahora veremos una forma un poco más sencilla de mostrar estas cónicas con la ayuda de una nueva biblioteca **sympy**.

```python
import sympy as sp
from sympy.plotting import plot_implicit
from sympy.abc import x, y

# 1. Graficar Círculo: x^2 + y^2 = r^2
r = 3
circle = sp.Eq(x**2 + y**2, r**2)
plot_implicit(circle, (x, -r, r), (y, -r, r), title="Círculo: $x^2 + y^2 = r^2$")


```

```python
# 2. Graficar Elipse: x^2 / a^2 + y^2 / b^2 = 1
a, b = 4, 2
ellipse = sp.Eq(x**2 / a**2 + y**2 / b**2, 1)
plot_implicit(ellipse, (x, -a, a), (y, -b, b), title=f"Elipse: $\\frac{{x^2}}{{ {a}^2}} + \\frac{{y^2}}{{{b}^2}} = 1$")
```

```python

# 3. Graficar Parábola: y = a * x^2 + b * x + c
a_p, b_p, c_p = 1, 0, 0
parabola = sp.Eq(y, a_p * x**2 + b_p * x + c_p)
plot_implicit(parabola, (x, -5, 5), (y, -5, 5), title=f"Parábola: $y = {a_p}x^2 + {b_p}x + {c_p}$")

```

```python
# 4. Graficar Hipérbola: x^2 / a^2 - y^2 / b^2 = 1
a_h, b_h = 2, 1
hyperbola = sp.Eq(x**2 / a_h**2 - y**2 / b_h**2, 1)
plot_implicit(hyperbola, (x, -5, 5), (y, -5, 5), title=f"Hipérbola: $\\frac{{x^2}}{{{a_h}^2}} - \\frac{{y^2}}{{{b_h}^2}} = 1$")
```
