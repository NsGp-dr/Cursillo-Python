## Cargue del conjunto de Datos

Vamos a utilizar el conjunto de datos `Iris`, que contiene información sobre tres especies de flores Iris y sus características. Este dataset es ideal para realizar un análisis exploratorio de datos.

# **Análisis Univariado**

En esta sección, vamos a analizar las variables de nuestro conjunto de datos de forma individual. Esto nos ayudará a entender mejor la distribución y características de cada variable.



## ¿Qué es una Variable Categórica?

Una **variable categórica** es aquella que toma un número limitado de valores distintos y discretos. Estos valores representan categorías o etiquetas, y no tienen un orden inherente. Por ejemplo, en un conjunto de datos sobre automóviles, la variable "Tipo de combustible" podría ser categórica, con categorías como "Gasolina", "Diesel" o "Eléctrico".

### Ejercicio: Análisis de una Variable Categórica

Vamos a visualizar la distribución de una variable categórica utilizando un gráfico de barras. Esto nos permitirá ver cuántos datos pertenecen a cada categoría.


```python
# Reemplaza 'nombre_columna_categórica' con la columna de tu elección
sns.countplot(x='nombre_columna_categórica', data=df)
plt.title('Distribución de la Variable Categórica')
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.show()
```

## ¿Qué es una Variable Numérica?

Una variable numérica es aquella que toma valores numéricos y continuos. Estos valores pueden representar cantidades, mediciones o puntuaciones, y suelen tener un orden natural. Por ejemplo, en un conjunto de datos de estudiantes, la variable "Edad" sería numérica, ya que toma valores como 15, 16, 17, etc.
### **Ejercicio**: Análisis de una Variable Numérica
Para las variables numéricas, podemos usar histogramas y boxplots para entender cómo están distribuidos los datos. Los histogramas nos muestran la distribución de frecuencias, mientras que los boxplots nos ayudan a identificar posibles valores atípicos.\
**Histograma**\
Un histograma nos muestra la distribución de una variable numérica dividiendo los datos en intervalos y contando cuántos datos caen en cada intervalo.

```python
# Reemplaza 'nombre_columna_numérica' con la columna de tu elección
sns.histplot(df['nombre_columna_numérica'], kde=True)
plt.title('Distribución de la Variable Numérica')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Los cuantiles suelen usarse por grupos que dividen la distribución en partes iguales; entendidas estas como intervalos que comprenden la misma proporción de valores. Los más usados son:

  * Los cuartiles, que dividen a la distribución en cuatro partes (corresponden a los cuantiles 0,25; 0,50 y 0,75);
  * Los quintiles, que dividen a la distribución en cinco partes (corresponden a los cuantiles 0,20; 0,40; 0,60 y 0,80);
  * Los deciles, que dividen a la distribución en diez partes;
  * Los percentiles, que dividen a la distribución en cien partes.

# **Análisis Bivariado**

El análisis bivariado nos permite examinar cómo se relacionan dos variables entre sí. Este tipo de análisis es crucial para identificar patrones, tendencias, o relaciones significativas entre diferentes variables en nuestro conjunto de datos.

## Gráficos de Dispersión

Un **gráfico de dispersión** (scatter plot) nos ayuda a visualizar la relación entre dos variables numéricas. En este gráfico, cada punto representa una observación, con su posición determinada por los valores de las dos variables. Es útil para detectar correlaciones y tendencias lineales o no lineales entre las variables.

# **Ejercicio**: Creación de un Gráfico de Dispersión

Vamos a crear un gráfico de dispersión para ver cómo se relacionan dos variables numéricas.


```python
# Reemplaza 'var1' y 'var2' con las columnas numéricas que quieras analizar
sns.scatterplot(x='var1', y='var2', data=df)
#Podemos agregar el color de la variable categórica que querramos agregando esto al código anterior:
#sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=df, c=df["species"], cmap="viridis")
plt.title('Gráfico de Dispersión entre var1 y var2')
plt.xlabel('var1')
plt.ylabel('var2')
plt.show()
```

## **Matriz de Correlación**

Una matriz de correlación es una tabla que muestra las correlaciones entre todas las variables numéricas de nuestro conjunto de datos. La correlación mide la fuerza y dirección de la relación lineal entre dos variables. Los valores de correlación varían entre -1 y 1, donde:

    1 indica una correlación positiva perfecta.
    -1 indica una correlación negativa perfecta.
    0 indica que no hay correlación lineal.

### **Ejercicio**: Creación de una Matriz de Correlación

Vamos a calcular y visualizar la matriz de correlación para nuestras variables numéricas.

```python
# Calcular la matriz de correlación
correlation_matrix = df.corr()

# Visualizar la matriz de correlación
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

# **Detección de Valores Atípicos**

Los valores atípicos son observaciones que se alejan significativamente del resto de los datos. Identificarlos es importante porque pueden influir en el análisis y dar lugar a conclusiones incorrectas. Podemos detectar valores atípicos utilizando diferentes métodos, como los boxplots o el z-score.

## **Ejercicio**: Detección de Valores Atípicos con Boxplot y Z-score
**Detección de Outliers con Boxplot**

Vamos a utilizar un boxplot para visualizar los posibles valores atípicos en una variable numérica.

```python
# Reemplaza 'nombre_columna_numérica' con la columna de tu elección
sns.boxplot(x='sepal width (cm)', data=df)
plt.title('Boxplot para Detección de Valores Atípicos')
plt.xlabel('Valores')
plt.show()
```

**Detección de Outliers con Z-score**

El z-score mide cuántas desviaciones estándar se encuentra una observación del promedio. Un z-score muy alto o bajo puede indicar un valor atípico.

```python
from scipy import stats

# Calcular el z-score para una columna específica
df['z_score'] = stats.zscore(df['sepal width (cm)'])

# Filtrar los outliers (z-score > 3 o < -3)
outliers = df[(df['z_score'] > 3) | (df['z_score'] < -3)]
outliers
```
